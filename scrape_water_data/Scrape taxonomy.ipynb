```python
count = 0
taxon_list= []
for name in fish_list:
    count +=1
    try:
        fish =requests.get('https://en.wikipedia.org/wiki/' + name)
        time.sleep(2)
        fish_soup = BeautifulSoup(fish.text)

        info = fish_soup.find(class_ = 'biota').find_all('td')[3:17]
        print(name)
        taxon = [re.sub(r'\xa0','',x.get_text(strip=True)) for x in info]
        print(taxon)
        
        keys = [i for i in taxon if taxon.index(i)%2 ==0]
        values = [i for i in taxon if taxon.index(i)%2 !=0]
        taxon_dict = dict(zip(keys,values))
        taxon_dict['Name'] = name
        taxon_list.append(taxon_dict)
        
        
        
    except:
        print(name)
    if count > 10:
        break
```
