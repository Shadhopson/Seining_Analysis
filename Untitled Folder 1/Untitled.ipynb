```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import dateutil
```

```python
df = pd.read_csv('../Created CSVs/Cleaned_For_Analysis_All_Data.csv')
```

 1, Graph of species richness per year 2008-2019
 (" # of distinct species" on the y axis, "year" on x axis, Title: Species Richness per Year, make all bars a dark blue )

```python
df['Year'] = [x[:4] for x in df.Date]
```

```python
spec_rich_df = df[['Year', 'Corrected Common Name']]
```

```python
grouped_spec_rich_df = df[['Year', 'Grouped Corrected Common Name']]
```

```python
species_count_per_year  = {}
grouped_species_count_per_year  = {}
for year in range(2008,2020):
    species_count_per_year[year] = len(spec_rich_df[spec_rich_df['Year'] == str(year)]['Corrected Common Name'].unique())
    grouped_species_count_per_year[year] = len(grouped_spec_rich_df[grouped_spec_rich_df['Year'] == str(year)]['Grouped Corrected Common Name'].unique())
```

```python
plt.figure(figsize=(10,5))
sns.barplot(list(grouped_species_count_per_year.keys()), list(grouped_species_count_per_year.values()), color="#191970")
plt.xlabel('year')
plt.ylabel("# of distinct species")
plt.title("Species Richness Per Year")
plt.show()
```
