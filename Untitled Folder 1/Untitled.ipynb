2. bar graph of CPU 2016-2019
Looking at FISH ONLY. 
The y axis being average catch per unit effort (unit being 1 haul) so figuring out the average catch per haul for each date, then finding average for each year using all dates' averages. x axis is year.


```python
is_fish =pd.read_csv('is_it_a_fish.csv', header=None).fillna(0)
```

```python
is_fish_dict = dict(zip(is_fish[0], is_fish[1]))
```

```python
df['is_fish'] = [is_fish_dict[x] for x in df['Grouped Corrected Common Name']]
```

```python
cpu_df = df[['Grouped Corrected Common Name', "Corrected Quantity", "Corrected # Hauls", "Year", "Date", "is_fish"]]
```

```python
cpu_fish_df = cpu_df[cpu_df['is_fish'] == 1]
```

```python
cpu_fish_df = cpu_fish_df.dropna()
```

```python
cpu_graph_db = cpu_fish_df.groupby(['Date', 'Year', 'Corrected # Hauls'], as_index=False).sum()
```

```python
cpu_graph_db['Catch Per Unit Effort'] = cpu_graph_db['Corrected Quantity']/ cpu_graph_db['Corrected # Hauls']
```

```python
cpu_final = cpu_graph_db.groupby('Year', as_index= False).mean()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(cpu_final['Year'], cpu_final['Catch Per Unit Effort'], color="#191970")
plt.xlabel('Year')
plt.ylabel("Average Number Of Fish Caught Per Haul")
plt.title("Catch Per Unit Effort 2016-2019")
plt.savefig('Catch_Per_Unit_Effort_2016_2019.jpg')
```

 3. Stacked bar chart organisms caught per year (2011-2019 only)
switch the axises on the graph you already have made... years on the y, number organisms on the x. I would like to also have the exact number of each written within the colored bar of each, i.e if in 2016 there were 200 silversides, there would be a small 200 overlaying the yellow silverside section of the 2016 bar.. and etc for every single bar stack. the key needs to be off to the right side of the grph, ideally with a box around it and title that" Key" not species..

```python
stacked_df = df[['Year', 'Grouped Corrected Common Name', 'Corrected Quantity', 'is_fish']]
```

```python
stacked_df = stacked_df[stacked_df['Year'].isin(['2011','2012','2013', '2014', '2015','2016', '2017','2018','2019']) ]
```

```python
new_li = []
for index, row in stacked_df.iterrows():
    if row['Grouped Corrected Common Name'] not in ['atlantic silverside','comb jelly', 'shrimp', 'tunicates', 'sea snail']:
        if row['is_fish']:
            new_li.append('other fish')
        else: 
            new_li.append('other invertebrates')
        
    else:
        new_li.append(row['Grouped Corrected Common Name'])
```

```python
stacked_df['Categorized Common Name'] = new_li
```

```python
stacked_df = stacked_df.groupby(['Year', 'Categorized Common Name'], as_index=False).sum()
```

```python
animal_dict = {'other invertebrates':[],'other fish':[], 'shrimp': [], 'atlantic silverside':[],'comb jelly': [], 'tunicates': [], 'sea snail' : [] }
```

```python
for year in range(2011,2020):
    curr_df  = stacked_df[stacked_df['Year'] == str(year)]
    for animal in animal_dict:
        if len(curr_df[curr_df['Categorized Common Name'] == animal]['Corrected Quantity']) < 1:
            animal_dict[animal] += [0]
        else:
            animal_dict[animal] += list(curr_df[curr_df['Categorized Common Name'] == animal]['Corrected Quantity'])
```

```python
for animal in animal_dict:
    animal_dict[animal].insert(0,animal)
```

```python
head = ['Key'] + list(range(2011,2020))
```

```python
df2 = pd.DataFrame(columns=head, data=[animal_dict['other invertebrates'],animal_dict['other fish'], animal_dict['sea snail'],animal_dict['tunicates'], animal_dict['shrimp'], animal_dict['atlantic silverside'], animal_dict['comb jelly']])
```

```python
sns.set(rc={'figure.figsize':(10,8)})
df2.set_index('Key').T.plot(kind='bar', stacked=True)
plt.title('Number of Organisms Caught by Year')
plt.xlabel('Year')
plt.ylabel('Number of Organisms Caught')
plt.show()
```
