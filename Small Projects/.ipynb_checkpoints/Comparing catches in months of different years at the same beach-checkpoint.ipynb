```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('../Created CSVs/Cleaned_For_Analysis_All_Data.csv')
```

```python
df = df[~df['Date'].isin(['2018-05-30', '2018-06-02'])]
```

```python
df['Date'] =pd.to_datetime(df.Date)
```

```python
df['Year'] = [x.year for x in df.Date]
```

```python
df =df[df['Year'] >=2018]
```

```python
df['Month'] =  [x.month for x in df['Date']]
```

```python
gdf = df[['Month','Year','Corrected Common Name', 'Corrected Quantity']].groupby(['Month', 'Year', 'Corrected Common Name']).sum()
```

```python
#gdf.to_csv("Year_comparison_at_pier_4.csv")
```

```python
seines_per_month = df.groupby(['Month','Year'])['Date'].nunique()
```

```python
hauls_per_date = df[['Date','Month','Year', 'Corrected # Hauls']].groupby(['Date','Month','Year'] , as_index=False).mean()
```

```python
hauls_per_month = hauls_per_date.groupby(['Month','Year'])['Corrected # Hauls'].sum()
```

```python
hauls_per_month.replace(0, "No Data", inplace=True)
```

```python
grouped_count  =gdf.groupby(['Month','Year']).sum()
```

```python
grouped_count['# of Seines'] =seines_per_month
```

```python
grouped_count['# of Hauls'] = hauls_per_month
```

```python
grouped_count
```

```python
grouped_count['Corrected Quantity']
```

```python
ax = grouped_count['Corrected Quantity'].unstack().plot.bar(title ="Quantity Caught Per Month At Pier 4 Beach (2018 vs. 2019)" )\
.set_ylabel('Quantity Caught')

fig = ax.get_figure()
fig.savefig('Quantity Caught Per Month At Pier 4 Beach 2018 vs 2019')
```
