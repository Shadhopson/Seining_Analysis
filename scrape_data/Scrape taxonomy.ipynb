```python
def get_taxon(name):
        fish =requests.get('https://en.wikipedia.org/wiki/' + name)
        time.sleep(2)
        fish_soup = BeautifulSoup(fish.text)
        info = fish_soup.find(class_ = 'biota').find_all('td')
        taxon_dict = {}
        for i,x in enumerate(info):
            cx =  x.get_text(strip=True)
            if cx in ['Kingdom:', 'Phylum:','Class:','Order:','Family:','Genus:','Species:']:
                taxon_dict[cx] = re.sub(r'\xa0','',[x for x in info[i+1].stripped_strings][0])#info[i+1].get_text(strip=True))
        taxon_dict['Name'] = name
        return taxon_dict
    
```

```python

taxon_list= []
skip = False
for count,name in enumerate(fish_list):
    if skip == True:
        skip = False
        continue
    
    #if count < 148:
    #    continue
    try:
        taxon_list.append( get_taxon(name))
        
        
    except:
        if count + 1 > len(fish_list)-1:
            continue
        print(name, fish_list[count+1])
        taxon_dict = get_taxon(fish_list[count+1])
        taxon_dict['Name']= name
        taxon_list.append(taxon_dict)
        skip = True
        
        
    #if count > 168:
    #    break
```
