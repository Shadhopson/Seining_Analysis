```python
import pandas as pd
```

```python
fish_df = pd.read_csv('Name_Corrected_All_Seins.csv')
```

```python
fish_df.head()
```

```python
fish_names = set(list(fish_df['Species']))
```

```python
fish_df['Date'] = pd.to_datetime(fish_df['Date'])
```

```python
catch_by_year = fish_df[['Date', 'Species']].groupby(fish_df['Date'].map(lambda x: x.year))
```

```python
year_column = {}
catch_by_year.get_group(2008)['Species']
for group in catch_by_year:
    year_column[group[0]] = []
    for name in sorted(list(fish_names)):
        if name in set(list(group[1]['Species'])):
            year_column[group[0]].append('Yes')
        else:
            year_column[group[0]].append('No')
    #year_column.append(species)
```

```python
year_column
```

```python
fish_check_list = pd.DataFrame(data = sorted(list(fish_names)), columns = ['Species'] )
```

```python
years = range(2008,2018)
```

```python
for year in years:
    fish_check_list[year] =  year_column[year]
```

```python
fish_check_list
```

```python
def color_negative_red(val):
    """
    Takes a scalar and returns a string with
    the css property `'color: red'` for negative
    strings, black otherwise.
    """
    color = 'red' if val == 'No' else 'black'
    return 'color: %s' % color
```

```python
s = fish_check_list.style.applymap(color_negative_red)
```

```python
s
```

```python
def highlight_max(s):
    '''
    highlight the maximum in a Series yellow.
    '''
    is_max = s == 'Yes'
    return ['background-color: yellow' if v else '' for v in is_max]
```

```python
fish_check_list.style.apply(highlight_max)
```

```python
fish_check_list.to_csv('Species_Check_List_By_Year.csv')
```
