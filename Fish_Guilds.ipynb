```python
import pandas as pd
```

```python
DIRECTORY = 'Created CSVs/'
```

```python
name_df = pd.read_csv(DIRECTORY +'2017_CPUE_spp_handout.csv')
```

```python
data_graph = name_df.dropna().groupby('Spatial Guild').sum()
```

```python
groups  = name_df.dropna().groupby('Spatial Guild').sum()
```

```python
transposegroups = groups.transpose()
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
graph = groups.iloc[0].plot(title = 'Hard Benthic' )
graph.set_xlabel('Year')
graph.set_ylabel('Count')
```

```python
guilds = ['Hard-Benthic','Pelagic', "Proximo-Benthic", "Sheltered Shallow", "Soft Benthic","Weed"]
f, axarr = plt.subplots(6, sharex=True)
 
f.set_figwidth(6)
f.set_figheight(10)

for i in range(6):
    axarr[i].get_xaxis().get_major_formatter().set_useOffset(False)
    axarr[i].plot(groups.iloc[i])
    axarr[i].set_title(guilds[i])
    axarr[i].set_ylabel("Fish Caught") 
axarr[i].set_xlabel("Year")
```

```python
f.savefig("Graphs/Fish_guilds_per_year")
```
